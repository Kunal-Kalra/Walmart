###  Find the outliers and replace the outliers

##### Quantile based flooring and capping

```python
data.head()
```

```python
sns.boxplot(x = data.age)
```

```python
age_10_percentile = np.percentile(data.age, 10)
age_90_percentile = np.percentile(data.age, 90)
print(age_10_percentile, age_90_percentile)
```

```python
a = np.where(data.age<age_10_percentile, age_10_percentile, data.age)
a = np.where(a>age_90_percentile, age_90_percentile, a)
data.age = a
```

```python
sns.boxplot(x = data.age)
```

```python
sns.boxplot(x = data.fare)
```

```python
fare_10_percentile = np.percentile(data.fare, 10)
fare_90_percentile = np.percentile(data.fare, 90)
print(fare_10_percentile, fare_90_percentile)
```

```python
f = np.where(data.fare<fare_10_percentile, fare_10_percentile, data.fare)
f = np.where(f>fare_90_percentile, fare_90_percentile, f)
data.fare = f
```

```python
sns.boxplot(x= data.fare)
```

### Encoding Categorical Columns

```python
data.head()
```

```python
from sklearn.preprocessing import LabelEncoder
lr = LabelEncoder()
```

```python
data.sex = lr.fit_transform(data.sex)
data.adult_male = lr.fit_transform(data.adult_male)
data.alive = lr.fit_transform(data.alive)
data['class'] = lr.fit_transform(data['class'])
data.alone = lr.fit_transform(data.alone)
```

```python
data.embarked = lr.fit_transform(data.embarked)
data.who = lr.fit_transform(data.who)
```

```python
data.head()
```

```python
data = data.drop(columns=['class','alive'], axis=1)
#same columns removed
```

```python
data.head()
```

```python
data = pd.get_dummies(data, columns = ['embark_town'])
data.head()
```
