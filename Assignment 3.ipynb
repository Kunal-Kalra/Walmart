## Finding the outliers and replacing them 

```python
plt.figure(figsize = (10,20))
plt.subplot(3,2,1)
sns.boxplot(x=data['price'])
plt.subplot(3,2,2)
sns.boxplot(x=data['area'])
plt.subplot(3,2,3)
sns.boxplot(x=data['bedrooms'])
plt.subplot(3,2,4)
sns.boxplot(x=data['bathrooms'])
plt.subplot(3,2,5)
sns.boxplot(x=data['stories'])
plt.subplot(3,2,6)
sns.boxplot(x=data['parking'])
```

```python
price_10_percentile = np.percentile(data.price, 10)
price_90_percentile = np.percentile(data.price, 90)
print(price_10_percentile, price_90_percentile)
p = np.where(data.price<price_10_percentile, price_10_percentile, data.price)
p= np.where(p>price_90_percentile, price_90_percentile, p)
data.price = p
sns.boxplot(x = data.price)
```

```python
area_10_percentile = np.percentile(data.area, 10)
area_90_percentile = np.percentile(data.area, 90)
print(area_10_percentile, area_90_percentile)
a = np.where(data.area<area_10_percentile, area_10_percentile, data.area)
a= np.where(a>area_90_percentile, area_90_percentile, a)
data.area = a
sns.boxplot(x = data.area)
```

## Encoding

```python
from sklearn.preprocessing import LabelEncoder
lr = LabelEncoder()
```

```python
data.head()
```

```python
data.mainroad = lr.fit_transform(data.mainroad)
data.guestroom = lr.fit_transform(data.guestroom)
data.basement = lr.fit_transform(data.basement)
data.hotwaterheating = lr.fit_transform(data.hotwaterheating)
data.airconditioning = lr.fit_transform(data.airconditioning)
data.furnishingstatus = lr.fit_transform(data.furnishingstatus)
```

```python
data.head()
```

### Split the data into dependent and independent variables

```python
#Independent
x = data.drop(columns=['price'], axis=1)
#Dependent
y = data.price
```

### Scale the Independent Variables

```python
from sklearn.preprocessing import MinMaxScaler
scale = MinMaxScaler()
```

```python
a = scale.fit_transform(x)
x = pd.DataFrame(a, columns = x.columns)
x.head()
```

### Split the data into training and testing

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest, ytrain, ytest = train_test_split(x, y, test_size=0.25, random_state = 0)
```

## Build the Model

```python
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
```

```python
lr.fit(xtrain, ytrain)
```

```python
ypred = lr.predict(xtest)
```

```python
ypred
```

## Measuring the performance using Metrics

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
print("R2 score =", r2_score(ytest, ypred))
print("MSE =",mean_squared_error(ytest, ypred))
print("RMSE =", (mean_squared_error(ytest, ypred))**0.5)
print("MAE =",mean_absolute_error(ytest, ypred))
```
